# CCUS Emissions Simulator (2025–2050)

This notebook simulates global greenhouse gas emissions with basic **CCUS (carbon capture, utilization, and storage)**.
We will:
- Start with **gross global emissions** at 35,000 Mt/year and let them **decline slightly each year**.
- Start **CCUS** at 40 Mt/year and let it **grow by a fixed % each year**.
- Compute **net emissions = gross − captured**.
- Plot **gross, captured, and net** from **2025 to 2050**.
- Provide a small function so you can **test different scenarios** (e.g., decline rate and CCUS growth).

**How to read this notebook (especially if you're new to Python):**
- Every **code cell starts** with a short plain-English explanation of what the cell does.
- **Every single line of code** has a comment starting with `# note:` explaining what that line does.




```python

# Plain-English explanation:
# This cell imports the Python libraries we will use (math, pandas, numpy, matplotlib).
# These tools help us handle numbers, tables, and charts.
# If you're very new: 'import' means "bring this toolbox into the code so we can use it."

import math  # note: provides math utilities (not strictly required here, but often handy)
import numpy as np  # note: numpy helps with efficient numeric arrays and calculations
import pandas as pd  # note: pandas helps create and manipulate data tables (DataFrame)
import matplotlib.pyplot as plt  # note: matplotlib creates charts/plots
from dataclasses import dataclass  # note: dataclass helps define simple parameter containers
```

## Core assumptions (you can change these later)

- **Years:** 2025 to 2050 (inclusive).
- **Gross emissions (start):** 35,000 Mt in 2025.
- **Gross decline rate:** a small % decline per year (default: 0.7%).
- **CCUS captured (start):** 40 Mt in 2025.
- **CCUS growth rate:** % increase per year (default: 15%).

> Tip: A 0.7% decline means multiply by 0.993 each year (`1 - 0.007`). A 15% growth means multiply by `1.15` each year.


```python

# Plain-English explanation:
# We define a small 'Scenario' container to hold model settings (like rates and starting values).
# Then we define a 'simulate_emissions' function that:
#   1) Builds a list of years (2025–2050)
#   2) Computes gross emissions each year with a steady decline
#   3) Computes CCUS captured each year with a steady growth
#   4) Computes net = gross − captured
#   5) Packages everything into a pandas DataFrame you can easily inspect/plot

@dataclass  # note: use dataclass to automatically create an initializer and nice representation
class Scenario:  # note: define a simple structure to store our scenario parameters
    start_year: int = 2025  # note: first simulation year
    end_year: int = 2050  # note: last simulation year (inclusive)
    gross_start_mt: float = 35000.0  # note: initial gross emissions in Mt for start_year
    gross_decline_rate_pct: float = 0.7  # note: percent per year decline of gross emissions
    ccus_start_mt: float = 40.0  # note: initial CCUS captured in Mt for start_year
    ccus_growth_rate_pct: float = 15.0  # note: percent per year growth of CCUS captured

def simulate_emissions(scn: Scenario) -> pd.DataFrame:
    """
    Build the simulation table for the given scenario.
    Returns a DataFrame with columns: Year, Gross_Mt, Captured_Mt, Net_Mt.
    """  # note: function docstring that explains purpose and outputs
    
    years = list(range(scn.start_year, scn.end_year + 1))  # note: create a list of integer years from start to end inclusive
    
    gross = []  # note: will hold yearly gross emissions values
    captured = []  # note: will hold yearly CCUS captured values
    
    gross_current = scn.gross_start_mt  # note: starting gross emissions level
    captured_current = scn.ccus_start_mt  # note: starting CCUS captured level
    
    gross_decline_factor = 1.0 - (scn.gross_decline_rate_pct / 100.0)  # note: convert percent decline to a multiplicative factor
    ccus_growth_factor = 1.0 + (scn.ccus_growth_rate_pct / 100.0)  # note: convert percent growth to a multiplicative factor
    
    for i, y in enumerate(years):  # note: loop over years with index i and value y
        if i == 0:  # note: for the first year, we use the starting values directly
            gross.append(gross_current)  # note: record starting gross
            captured.append(captured_current)  # note: record starting captured
        else:  # note: for subsequent years, apply decline/growth factors
            gross_current *= gross_decline_factor  # note: apply one year of gross decline
            captured_current *= ccus_growth_factor  # note: apply one year of CCUS growth
            gross.append(gross_current)  # note: record updated gross
            captured.append(captured_current)  # note: record updated captured
    
    gross_arr = np.array(gross, dtype=float)  # note: convert to numpy array for easy math
    captured_arr = np.array(captured, dtype=float)  # note: convert to numpy array
    net_arr = gross_arr - captured_arr  # note: compute net emissions element-wise
    
    df = pd.DataFrame({  # note: build a DataFrame from a dictionary of columns
        "Year": years,  # note: simulation years
        "Gross_Mt": gross_arr,  # note: gross emissions column (Mt)
        "Captured_Mt": captured_arr,  # note: captured via CCUS (Mt)
        "Net_Mt": net_arr,  # note: net emissions = gross − captured (Mt)
    })
    
    return df  # note: return the results table
```

```python

# Plain-English explanation:
# Here we create a default scenario and run the simulation.
# We then show the first few rows so you can see the structure.

default_scn = Scenario(  # note: instantiate Scenario with defaults (you can tweak values here)
    start_year=2025,  # note: start simulation year
    end_year=2050,  # note: end simulation year
    gross_start_mt=35000.0,  # note: starting gross emissions in Mt
    gross_decline_rate_pct=0.7,  # note: gross declines 0.7% per year
    ccus_start_mt=40.0,  # note: starting captured amount in Mt
    ccus_growth_rate_pct=15.0  # note: captured grows 15% per year
)

df_default = simulate_emissions(default_scn)  # note: run the simulation for the default scenario
df_default.head()  # note: display the first few rows of the DataFrame
```

```python

# Plain-English explanation:
# This cell plots Gross, Captured, and Net emissions on the same chart.
# We use matplotlib for plotting. We don't set any specific colors so your theme decides.
# After running, scroll below the cell to see the chart.

plt.figure()  # note: create a new figure for the plot
plt.plot(df_default["Year"], df_default["Gross_Mt"], label="Gross (Mt)")  # note: plot gross over time
plt.plot(df_default["Year"], df_default["Captured_Mt"], label="Captured via CCUS (Mt)")  # note: plot captured over time
plt.plot(df_default["Year"], df_default["Net_Mt"], label="Net = Gross − Captured (Mt)")  # note: plot net over time
plt.xlabel("Year")  # note: label the x-axis
plt.ylabel("Emissions (Mt CO₂e)")  # note: label the y-axis
plt.title("Emissions Path with CCUS (Default Scenario)")  # note: set the plot title
plt.legend()  # note: show legend so lines are labeled
plt.grid(True)  # note: add a grid for readability
plt.show()  # note: render the plot
```

```python

# Plain-English explanation:
# Let's compute and print some quick summary numbers:
# - Gross, Captured, and Net in the final year
# - Total (cumulative) captured over the whole period

final_row = df_default.iloc[-1]  # note: select the last row (year 2050)
final_year = int(final_row["Year"])  # note: extract final year as integer
final_gross = float(final_row["Gross_Mt"])  # note: final gross emissions
final_captured = float(final_row["Captured_Mt"])  # note: final captured amount
final_net = float(final_row["Net_Mt"])  # note: final net emissions
total_captured = float(df_default["Captured_Mt"].sum())  # note: sum of captured across all years

print(f"Final year: {final_year}")  # note: print final year
print(f"Gross in {final_year}: {final_gross:,.1f} Mt")  # note: formatted gross
print(f"Captured in {final_year}: {final_captured:,.1f} Mt")  # note: formatted captured
print(f"Net in {final_year}: {final_net:,.1f} Mt")  # note: formatted net
print(f"Total captured (2025–{final_year}): {total_captured:,.1f} Mt")  # note: total captured over full period
```

```python

# Plain-English explanation:
# This helper function lets you quickly test different scenarios without editing earlier cells.
# Provide your own decline rate (% per year) and CCUS growth rate (% per year), plus optional start values.
# It will run the simulation, print the final-year numbers, and plot the results.

def run_scenario(gross_decline_rate_pct=0.7, ccus_growth_rate_pct=15.0,
                 gross_start_mt=35000.0, ccus_start_mt=40.0,
                 start_year=2025, end_year=2050):
    # note: define a function with default parameters for easy experimentation
    
    scn = Scenario(  # note: build a Scenario with your chosen parameters
        start_year=start_year,  # note: use provided start year
        end_year=end_year,  # note: use provided end year
        gross_start_mt=gross_start_mt,  # note: starting gross emissions
        gross_decline_rate_pct=gross_decline_rate_pct,  # note: yearly gross decline rate (%)
        ccus_start_mt=ccus_start_mt,  # note: starting CCUS captured
        ccus_growth_rate_pct=ccus_growth_rate_pct  # note: yearly CCUS growth rate (%)
    )
    
    df = simulate_emissions(scn)  # note: run the simulation for this scenario
    
    final = df.iloc[-1]  # note: grab last row for final-year values
    print(f"Final year: {int(final['Year'])}")  # note: print final year
    print(f"Gross: {float(final['Gross_Mt']):,.1f} Mt")  # note: print final gross
    print(f"Captured: {float(final['Captured_Mt']):,.1f} Mt")  # note: print final captured
    print(f"Net: {float(final['Net_Mt']):,.1f} Mt")  # note: print final net
    
    plt.figure()  # note: open a new figure for plotting
    plt.plot(df["Year"], df["Gross_Mt"], label="Gross (Mt)")  # note: plot gross
    plt.plot(df["Year"], df["Captured_Mt"], label="Captured via CCUS (Mt)")  # note: plot captured
    plt.plot(df["Year"], df["Net_Mt"], label="Net = Gross − Captured (Mt)")  # note: plot net
    plt.xlabel("Year")  # note: x-axis label
    plt.ylabel("Emissions (Mt CO₂e)")  # note: y-axis label
    plt.title("Emissions Path with CCUS — Custom Scenario")  # note: plot title
    plt.legend()  # note: show legend
    plt.grid(True)  # note: add grid
    plt.show()  # note: display the plot
    
    return df  # note: return the scenario results in case you want to inspect or save
```

```python
# Plain-English explanation:
# This cell adds a simple "stock and flow" model for atmospheric CO₂ (in ppm).
# - We take your table of yearly NET emissions (Mt CO₂ per year).
# - Convert yearly net emissions to ppm (using 1 ppm ≈ 7.8 Gt CO₂).
# - Update an atmospheric stock: Stock_next = Stock_now + Emissions_ppm - SinkRemoval.
# - SinkRemoval is a simple linear sink: k * (Stock_now - preindustrial_ppm).
# You can tweak: starting_ppm, preindustrial_ppm, and k (removal rate per year).

import numpy as np  # note: ensure numpy is available (already imported earlier)
import pandas as pd  # note: ensure pandas is available (already imported earlier)
import matplotlib.pyplot as plt  # note: plotting

# --- parameters you can tweak ---
preindustrial_ppm = 280.0  # note: a common preindustrial CO₂ baseline
starting_ppm = 420.0       # note: choose a modern starting level (adjust as you like)
ppm_per_GtCO2 = 1.0 / 7.8  # note: conversion from Gt CO₂ to ppm (≈0.128205)
k_sink = 0.012             # note: annual removal rate (~1.2%/yr) of excess above preindustrial

# --- use your existing results (df_default) ---
df = df_default.copy()  # note: take a copy so we don't mutate the original

# --- prepare arrays for the stock simulation ---
ppm_series = []  # note: will store atmospheric CO₂ (ppm) per year
ppm_now = starting_ppm  # note: initialize the stock at the chosen starting ppm

for _, row in df.iterrows():  # note: iterate over each year in the table
    net_Mt = float(row["Net_Mt"])  # note: get net emissions in Mt for this year
    net_Gt = net_Mt / 1000.0       # note: convert Mt -> Gt
    add_ppm = net_Gt * ppm_per_GtCO2  # note: convert Gt to ppm added to atmosphere
    
    excess_ppm = ppm_now - preindustrial_ppm  # note: how much above preindustrial
    sink_removal = k_sink * excess_ppm        # note: linear sink removes a fraction of excess
    
    ppm_next = ppm_now + add_ppm - sink_removal  # note: update the stock
    ppm_series.append(ppm_next)  # note: record the new stock
    ppm_now = ppm_next           # note: carry forward to next year

# --- attach to DataFrame and plot ---
df["Atmospheric_CO2_ppm"] = ppm_series  # note: add the new series to the table

# quick look
df[["Year", "Net_Mt", "Atmospheric_CO2_ppm"]].head()

# plot net emissions and CO₂ stock
plt.figure()  # note: new figure for emissions
plt.plot(df["Year"], df["Net_Mt"], label="Net Emissions (Mt/yr)")
plt.xlabel("Year")
plt.ylabel("Mt CO₂ per year")
plt.title("Net Emissions")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()  # note: new figure for atmospheric CO₂ stock
plt.plot(df["Year"], df["Atmospheric_CO2_ppm"], label="Atmospheric CO₂ (ppm)")
plt.xlabel("Year")
plt.ylabel("ppm")
plt.title("Atmospheric CO₂ Stock (Simple One-Box Sink)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plain-English explanation:
# This cell builds a more "interesting" agent-based model with behaviors:
# - Three agents (Power, Industry, Transport) each start with different emissions and CCUS.
# - Their decline/growth rates can change after certain years (policy triggers).
# - There are two shocks: a one-year recession (extra gross decline in 2032)
#   and a CCUS setback in 2043 (temporary 10% reduction in captured).
# - There's a tiny learning effect: if cumulative captured (global) passes a threshold,
#   CCUS growth for all agents speeds up a bit (simulating cost learning).
# - We simulate year-by-year, record per-agent and total Gross/Captured/Net, and plot.
#
# Tips:
# - Tweak parameters at the top to see different dynamics.
# - You can add more agents by copying the entries in the 'agents' list.

from dataclasses import dataclass  # note: dataclass helps define small "data containers"
import numpy as np  # note: numpy for easy numeric arrays
import pandas as pd  # note: pandas for tables
import matplotlib.pyplot as plt  # note: matplotlib for plotting

# -------- Parameters you can tweak --------
start_year = 2025  # note: first year
end_year   = 2050  # note: last year (inclusive)

# note: global learning threshold for CCUS (in Mt cumulative); crossing this boosts CCUS growth slightly
learning_threshold_mt = 5000.0  # note: when total captured since 2025 exceeds this, growth accelerates by +2 pp

# note: shocks (simple examples)
recession_year = 2032  # note: a one-year macro shock
extra_decline_this_year = 0.02  # note: in the recession year, gross gets an extra -2%

ccus_setback_year = 2043  # note: a one-year CCUS setback
ccus_temporary_cut = 0.10  # note: captured is cut by 10% that year

@dataclass  # note: define an Agent with behavior knobs
class Agent:
    name: str  # note: agent name for labeling
    gross_start_mt: float  # note: initial gross emissions in Mt
    ccus_start_mt: float  # note: initial captured in Mt
    gross_decline_pct_pre: float  # note: annual % gross decline before policy trigger
    gross_decline_pct_post: float  # note: annual % gross decline after policy trigger
    ccus_growth_pct_pre: float  # note: annual % CCUS growth before policy trigger
    ccus_growth_pct_post: float  # note: annual % CCUS growth after policy trigger
    policy_year: int  # note: year when agent switches from 'pre' to 'post' behavior
    ccus_cap_mt: float | None = None  # note: optional soft cap on CCUS level (None for no cap)

# -------- Define agents with different dynamics --------
agents = [
    Agent(
        name="Power",  # note: power sector
        gross_start_mt=15000.0,  # note: starts high
        ccus_start_mt=20.0,  # note: modest CCUS initially
        gross_decline_pct_pre=1.0,  # note: -1%/yr before policy year
        gross_decline_pct_post=2.5,  # note: -2.5%/yr after 2030 (accelerated decarbonization)
        ccus_growth_pct_pre=18.0,  # note: fast CCUS scale-up early
        ccus_growth_pct_post=22.0,  # note: even faster after policy trigger
        policy_year=2030,  # note: earlier push in power
        ccus_cap_mt=3000.0  # note: soft cap to avoid exploding to infinity
    ),
    Agent(
        name="Industry",  # note: heavy industry
        gross_start_mt=12000.0,  # note: substantial emissions
        ccus_start_mt=15.0,  # note: some CCUS
        gross_decline_pct_pre=0.6,  # note: slower early decline
        gross_decline_pct_post=1.5,  # note: improves after policy year
        ccus_growth_pct_pre=10.0,  # note: moderate early growth
        ccus_growth_pct_post=16.0,  # note: improves with policy support
        policy_year=2035,  # note: later policy kick-in
        ccus_cap_mt=2500.0  # note: optional cap
    ),
    Agent(
        name="Transport",  # note: transport sector
        gross_start_mt=8000.0,  # note: large but declining slowly
        ccus_start_mt=5.0,  # note: tiny CCUS initially
        gross_decline_pct_pre=0.3,  # note: slow early decline
        gross_decline_pct_post=2.0,  # note: speeds up with electrification
        ccus_growth_pct_pre=8.0,  # note: weak CCUS growth early
        ccus_growth_pct_post=12.0,  # note: modestly better later
        policy_year=2040,  # note: late acceleration
        ccus_cap_mt=600.0  # note: small cap (transport CCUS limited)
    ),
]

# -------- Simulation containers --------
years = list(range(start_year, end_year + 1))  # note: list of years
records = []  # note: will collect per-year, per-agent records (dicts)
cumulative_captured_mt = 0.0  # note: initialize global cumulative captured

# -------- Simulate year by year --------
# note: keep current state for each agent in a dict we can update
state = {
    a.name: {
        "gross": a.gross_start_mt,  # note: current gross
        "capt": a.ccus_start_mt,  # note: current captured
    } for a in agents
}

for y in years:  # note: iterate through each year
    # note: determine if learning bonus applies this year (based on last year's cumulative)
    learning_bonus_pp = 2.0 if cumulative_captured_mt >= learning_threshold_mt else 0.0  # note: add to CCUS growth %

    year_total_captured = 0.0  # note: track total captured this year for cumulative update

    for a in agents:  # note: update each agent
        st = state[a.name]  # note: get mutable state for this agent

        # --- choose behavior regime (pre vs post policy year) ---
        if y < a.policy_year:  # note: before policy trigger
            gross_decline_pct = a.gross_decline_pct_pre  # note: use pre-policy decline
            ccus_growth_pct = a.ccus_growth_pct_pre + learning_bonus_pp  # note: add learning if active
        else:  # note: after (or in) policy year
            gross_decline_pct = a.gross_decline_pct_post  # note: use post-policy decline
            ccus_growth_pct = a.ccus_growth_pct_post + learning_bonus_pp  # note: add learning if active

        # --- convert % to multiplicative factors ---
        gross_factor = 1.0 - gross_decline_pct / 100.0  # note: e.g., -1% => factor 0.99
        ccus_factor  = 1.0 + ccus_growth_pct / 100.0  # note: e.g., +18% => factor 1.18

        # --- apply recession shock (one year only) ---
        if y == recession_year:  # note: in the shock year, add extra gross decline
            gross_factor *= (1.0 - extra_decline_this_year)  # note: multiply by (1 - 0.02) => extra -2%

        # --- update levels (except first year uses start values we already hold) ---
        if y == start_year:  # note: first year uses initial values as-is
            gross_now = st["gross"]  # note: carry starting gross
            capt_now  = st["capt"]  # note: carry starting captured
        else:
            gross_now = st["gross"] * gross_factor  # note: update gross by decline factor
            capt_now  = st["capt"] * ccus_factor  # note: update captured by growth factor

        # --- apply one-year CCUS setback ---
        if y == ccus_setback_year:  # note: reduce captured for this year only
            capt_now *= (1.0 - ccus_temporary_cut)  # note: e.g., minus 10%

        # --- apply optional soft caps on CCUS ---
        if a.ccus_cap_mt is not None:  # note: only if cap is defined
            capt_now = min(capt_now, a.ccus_cap_mt)  # note: clamp to cap

        # --- compute net and record ---
        net_now = gross_now - capt_now  # note: net emissions
        records.append({  # note: one row per agent-year
            "Year": y,
            "Agent": a.name,
            "Gross_Mt": gross_now,
            "Captured_Mt": capt_now,
            "Net_Mt": net_now,
        })

        # --- store back state for next loop ---
        st["gross"] = gross_now  # note: update state
        st["capt"] = capt_now  # note: update state

        # --- update cumulative captured tracker ---
        year_total_captured += capt_now  # note: add to this year's total captured

    cumulative_captured_mt += year_total_captured  # note: update cumulative after all agents in this year

# -------- Build DataFrame and totals --------
df_abm = pd.DataFrame(records)  # note: convert list of dicts into a table

# note: pivot to wide per-agent for convenience (optional)
df_wide = df_abm.pivot(index="Year", columns="Agent", values="Net_Mt")  # note: net by agent as columns
df_wide = df_wide.reset_index()  # note: back to a normal frame

# note: totals by year
totals = df_abm.groupby("Year").agg(
    Gross_Total=("Gross_Mt", "sum"),
    Captured_Total=("Captured_Mt", "sum"),
    Net_Total=("Net_Mt", "sum"),
).reset_index()

# -------- Print a few interesting summaries --------
final = totals.iloc[-1]  # note: last year row
peak_captured = totals["Captured_Total"].max()  # note: maximum yearly captured over period
min_net = totals["Net_Total"].min()  # note: lowest net across period
print(f"Final year {int(final['Year'])}: Gross={final['Gross_Total']:,.0f} Mt, "
      f"Captured={final['Captured_Total']:,.0f} Mt, Net={final['Net_Total']:,.0f} Mt")
print(f"Max yearly captured: {peak_captured:,.0f} Mt")
print(f"Lowest yearly net (best): {min_net:,.0f} Mt")
print(f"Cumulative captured 2025–{int(final['Year'])}: {df_abm.groupby('Year')['Captured_Mt'].sum().sum():,.0f} Mt")

# -------- Plots --------
# Totals
plt.figure()
plt.plot(totals["Year"], totals["Gross_Total"], label="Gross Total (Mt)")
plt.plot(totals["Year"], totals["Captured_Total"], label="Captured Total (Mt)")
plt.plot(totals["Year"], totals["Net_Total"], label="Net Total (Mt)")
plt.xlabel("Year")
plt.ylabel("Mt CO₂e")
plt.title("Agent-Based (v2): Totals with Policies, Shocks, and Learning")
plt.grid(True)
plt.legend()
plt.show()

# Per-agent net (lines)
plt.figure()
for a in [ag.name for ag in agents]:  # note: loop agent names
    sub = df_abm[df_abm["Agent"] == a]  # note: rows for one agent
    plt.plot(sub["Year"], sub["Net_Mt"], label=f"Net {a} (Mt)")  # note: plot its net
plt.xlabel("Year")
plt.ylabel("Mt CO₂e")
plt.title("Agent-Based (v2): Per-Agent Net Emissions")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plain-English explanation:
# This cell builds ONE unified table that includes:
# - Year
# - Gross, Captured, Net emissions (from agent-based totals if available; otherwise df_default)
# - Cumulative Captured (Mt) and Cumulative Net (Gt)
# - Atmospheric CO₂ (ppm) computed as a stock with a simple sink
# It then plots ppm. Everything reuses data from earlier cells.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----- pick the emissions source (ABM totals preferred) -----
if "totals" in globals():  # note: if agent-based 'totals' DataFrame exists
    src = totals.copy()
    src.rename(columns={"Gross_Total": "Gross_Mt",
                        "Captured_Total": "Captured_Mt",
                        "Net_Total": "Net_Mt"}, inplace=True)
    source_used = "agent-based totals"
elif "df_default" in globals():  # note: fallback to simple single-scenario results
    src = df_default.copy()
    source_used = "default scenario"
else:
    raise RuntimeError("No emissions table found. Run the default simulation or the agent-based cell first.")

# ----- cumulative helpers -----
src = src.sort_values("Year").reset_index(drop=True)
src["Cum_Captured_Mt"] = src["Captured_Mt"].cumsum()
src["Cum_Net_Gt"] = (src["Net_Mt"] / 1000.0).cumsum()

# ----- ppm stock model -----
preindustrial_ppm = 280.0
starting_ppm = 420.0
ppm_per_GtCO2 = 1.0 / 7.8
k_sink = 0.012

net_gt = src["Net_Mt"].to_numpy() / 1000.0
add_ppm = net_gt * ppm_per_GtCO2

ppm_vals = []
ppm_now = starting_ppm
for i in range(len(src)):
    excess = ppm_now - preindustrial_ppm
    sink = k_sink * excess
    ppm_next = ppm_now + add_ppm[i] - sink
    ppm_vals.append(ppm_next)
    ppm_now = ppm_next

src["Atmospheric_CO2_ppm"] = ppm_vals

print(f"Using {source_used}. Unified table preview:")
print(src.head())
```

```python
# Plain-English explanation:
# This cell makes a twin-axis plot:
# - Left y-axis: Gross, Captured, Net emissions (Mt CO₂/year)
# - Right y-axis: Atmospheric CO₂ (ppm) from the stock model
# This lets you see annual emissions paths and atmospheric concentration together.

import matplotlib.pyplot as plt  # note: plotting (already imported above)

plt.figure(figsize=(8,5))  # note: create a new figure with size

ax1 = plt.gca()  # note: main axis (left y-axis)
ax2 = ax1.twinx()  # note: create a twin axis sharing the same x-axis but different y

# --- plot emissions on the left axis ---
ax1.plot(src["Year"], src["Gross_Mt"], label="Gross (Mt)", color="tab:blue")
ax1.plot(src["Year"], src["Captured_Mt"], label="Captured (Mt)", color="tab:orange")
ax1.plot(src["Year"], src["Net_Mt"], label="Net (Mt)", color="tab:green")
ax1.set_xlabel("Year")
ax1.set_ylabel("Emissions (Mt CO₂/year)")
ax1.grid(True)

# --- plot atmospheric CO₂ ppm on the right axis ---
ax2.plot(src["Year"], src["Atmospheric_CO2_ppm"], label="Atmospheric CO₂ (ppm)", color="tab:red")
ax2.set_ylabel("Atmospheric CO₂ (ppm)")

# --- combine legends from both axes ---
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper right")

plt.title("Emissions and Atmospheric CO₂ (ppm)")
plt.show()
```

```python
# Plain-English explanation:
# This cell creates a little GUI so you can change inputs with sliders instead of editing code.
# You can pick the data source (your default scenario or the agent-based totals if present),
# tweak gross decline and CCUS growth, starting values, and ppm stock settings.
# When you move a slider, it recomputes the results and draws a twin-axis chart:
#   - Left axis: Gross, Captured, Net (Mt/yr)
#   - Right axis: Atmospheric CO₂ (ppm)

import numpy as np  # note: numbers
import pandas as pd  # note: tables
import matplotlib.pyplot as plt  # note: plots
from ipywidgets import interact, FloatSlider, IntSlider, Dropdown, HBox, VBox, HTML  # note: GUI widgets

# --- tiny helper: build or fetch an emissions table with columns Year, Gross_Mt, Captured_Mt, Net_Mt ---
def _get_emissions_table(source_choice, 
                         gross_decline_rate_pct, ccus_growth_rate_pct,
                         gross_start_mt, ccus_start_mt,
                         start_year, end_year):
    # note: Prefer user's ABM totals if chosen and available
    if source_choice == "Agent-based totals (if available)" and "totals" in globals():
        src = totals.copy()  # note: use ABM totals
        src = src.rename(columns={"Gross_Total":"Gross_Mt","Captured_Total":"Captured_Mt","Net_Total":"Net_Mt"})
        src = src[["Year","Gross_Mt","Captured_Mt","Net_Mt"]].sort_values("Year").reset_index(drop=True)
        return src, "agent-based totals"
    # note: Otherwise (or if ABM not present), synthesize with current sliders via your core simulate_emissions
    if "Scenario" in globals() and "simulate_emissions" in globals():
        scn = Scenario(start_year=start_year, end_year=end_year,
                       gross_start_mt=gross_start_mt, gross_decline_rate_pct=gross_decline_rate_pct,
                       ccus_start_mt=ccus_start_mt, ccus_growth_rate_pct=ccus_growth_rate_pct)
        df = simulate_emissions(scn)
        return df[["Year","Gross_Mt","Captured_Mt","Net_Mt"]].copy(), "default scenario (live)"
    # note: If core funcs aren't defined, try to fall back to df_default if it exists
    if "df_default" in globals():
        return df_default[["Year","Gross_Mt","Captured_Mt","Net_Mt"]].copy(), "default scenario (precomputed)"
    raise RuntimeError("No emissions source found. Run earlier cells that define Scenario/simulate_emissions or ABM totals.")

# --- ppm stock update, returns ppm array ---
def _ppm_from_net(years, net_mt, starting_ppm, preindustrial_ppm, k_sink, ppm_per_GtCO2):
    net_gt = np.asarray(net_mt, dtype=float) / 1000.0  # note: Mt -> Gt
    add_ppm = net_gt * ppm_per_GtCO2  # note: emissions contribution to ppm before sink
    ppm_vals, ppm_now = [], float(starting_ppm)  # note: initialize
    for add in add_ppm:
        excess = ppm_now - preindustrial_ppm  # note: ppm above baseline
        sink = k_sink * excess  # note: simple linear sink
        ppm_next = ppm_now + add - sink  # note: update stock
        ppm_vals.append(ppm_next)  # note: record
        ppm_now = ppm_next  # note: carry forward
    return np.array(ppm_vals, dtype=float)

# --- GUI and plot ---
def _ui(source_choice, gross_decline_rate_pct, ccus_growth_rate_pct,
        gross_start_mt, ccus_start_mt, start_year, end_year,
        starting_ppm, preindustrial_ppm, k_sink):
    try:
        src, used = _get_emissions_table(source_choice, gross_decline_rate_pct, ccus_growth_rate_pct,
                                         gross_start_mt, ccus_start_mt, start_year, end_year)
    except Exception as e:
        print(f"Error getting emissions: {e}")
        return
    
    # note: compute ppm track
    ppm_per_GtCO2 = 1.0 / 7.8  # note: ~0.1282 ppm per Gt CO₂
    ppm_vals = _ppm_from_net(src["Year"].to_numpy(), src["Net_Mt"].to_numpy(),
                             starting_ppm, preindustrial_ppm, k_sink, ppm_per_GtCO2)
    src = src.copy()
    src["Atmospheric_CO2_ppm"] = ppm_vals  # note: attach ppm to table
    
    # --- twin-axis plot ---
    plt.figure(figsize=(8,5))  # note: new figure
    ax1 = plt.gca()  # note: left axis
    ax2 = ax1.twinx()  # note: right axis
    
    ax1.plot(src["Year"], src["Gross_Mt"], label="Gross (Mt)")  # note: left axis series
    ax1.plot(src["Year"], src["Captured_Mt"], label="Captured (Mt)")
    ax1.plot(src["Year"], src["Net_Mt"], label="Net (Mt)")
    ax1.set_xlabel("Year")
    ax1.set_ylabel("Emissions (Mt CO₂/year)")
    ax1.grid(True)
    
    ax2.plot(src["Year"], src["Atmospheric_CO2_ppm"], label="Atmospheric CO₂ (ppm)")  # note: right axis series
    ax2.set_ylabel("Atmospheric CO₂ (ppm)")
    
    # note: combined legend
    l1, lab1 = ax1.get_legend_handles_labels()
    l2, lab2 = ax2.get_legend_handles_labels()
    ax1.legend(l1 + l2, lab1 + lab2, loc="upper right")
    plt.title(f"Emissions & CO₂ ppm  — source: {used}")
    plt.show()
    
    # --- quick text readout (final year snapshot) ---
    last = src.iloc[-1]
    print(f"Final year {int(last['Year'])}: "
          f"Gross={last['Gross_Mt']:,.0f} Mt, "
          f"Captured={last['Captured_Mt']:,.0f} Mt, "
          f"Net={last['Net_Mt']:,.0f} Mt, "
          f"ppm={last['Atmospheric_CO2_ppm']:,.1f}")

# --- build the widgets ---
from ipywidgets import interact, FloatSlider, IntSlider, Dropdown, Layout

from ipywidgets import FloatSlider, IntSlider, Dropdown, Label, GridBox, Layout, interactive_output, VBox

# 1) Make widgets (no descriptions on sliders)
source_choice = Dropdown(
    options=["Agent-based totals (if available)", "Default scenario (live)"],
    value="Agent-based totals (if available)"
)
gross_decline_rate_pct = FloatSlider(min=0, max=5, step=0.1, value=0.7)
ccus_growth_rate_pct   = FloatSlider(min=0, max=40, step=0.5, value=15.0)
gross_start_mt         = IntSlider(min=10000, max=50000, step=500, value=35000)
ccus_start_mt          = IntSlider(min=0, max=200, step=5, value=40)
start_year             = IntSlider(min=2020, max=2030, step=1, value=2025)
end_year               = IntSlider(min=2030, max=2060, step=1, value=2050)
starting_ppm           = FloatSlider(min=350, max=500, step=1, value=420)
preindustrial_ppm      = FloatSlider(min=250, max=300, step=0.5, value=280)
k_sink                 = FloatSlider(min=0, max=0.05, step=0.001, value=0.012)

# 2) Lay them out in a two-column grid: Label | Control
grid = GridBox(
    children=[
        Label("Data source:"),         source_choice,
        Label("Gross decline %/yr:"),  gross_decline_rate_pct,
        Label("CCUS growth %/yr:"),    ccus_growth_rate_pct,
        Label("Gross start (Mt):"),    gross_start_mt,
        Label("CCUS start (Mt):"),     ccus_start_mt,
        Label("Start year:"),          start_year,
        Label("End year:"),            end_year,
        Label("Start ppm:"),           starting_ppm,
        Label("Preindustrial ppm:"),   preindustrial_ppm,
        Label("Sink rate /yr:"),       k_sink,
    ],
    layout=Layout(
        grid_template_columns="220px 520px",  # widen if needed
        grid_gap="6px 12px",
        width="800px"
    )
)

# 3) Wire widgets to your _ui() with interactive_output
out = interactive_output(
    _ui,
    {
        "source_choice": source_choice,
        "gross_decline_rate_pct": gross_decline_rate_pct,
        "ccus_growth_rate_pct": ccus_growth_rate_pct,
        "gross_start_mt": gross_start_mt,
        "ccus_start_mt": ccus_start_mt,
        "start_year": start_year,
        "end_year": end_year,
        "starting_ppm": starting_ppm,
        "preindustrial_ppm": preindustrial_ppm,
        "k_sink": k_sink,
    }
)

display(VBox([grid, out]))
```
