```python
# Plain-English explanation:
# This single cell defines a Scenario "data card", a simulate_emissions(...) function that
# computes Gross, Captured, Net *and* Atmospheric CO2 (ppm), runs a default scenario,
# and plots emissions (left axis) together with ppm (right axis).
import math  # For Mathsy sutff
import numpy as np  # For using numbers
import pandas as pd  # To create dataframes (df)
import matplotlib.pyplot as plt  # for plotting
from dataclasses import dataclass  # for defining parameter containers

# --- scenario "card" with defaults you can tweak ---
@dataclass
class Scenario:
    start_year: int = 2025                             # first year
    end_year: int = 2050                               # last year (inclusive)
    gross_start_mt: float = 35000.0                    # starting gross emissions (Mt/yr)
    gross_decline_rate_pct: float = 0.7                # gross declines % per year
    ccus_start_mt: float = 40.0                        # starting CCUS captured (Mt/yr)
    ccus_growth_rate_pct: float = 15.0                 # CCUS grows % per year

    # ppm stock-model parameters
    starting_ppm: float = 420.0                        # note: atmospheric CO2 at t0 (ppm)
    preindustrial_ppm: float = 280.0                   # baseline ppm to measure "excess"
    ppm_per_GtCO2: float = 1.0 / 7.8                   # ~0.1282 ppm per GtCO2 (rule of thumb)
    k_sink: float = 0.012                              # annual sink fraction of (ppm - preindustrial)

# --- simulator that ALSO computes ppm each year ---
def simulate_emissions(scn: Scenario) -> pd.DataFrame:
    """
    Build a table with Year, Gross_Mt, Captured_Mt, Net_Mt, Atmospheric_CO2_ppm.
    Gross declines geometrically; CCUS grows geometrically; ppm is a stock with a linear sink.
    """
    years = list(range(scn.start_year, scn.end_year + 1))                 # [2025, ..., 2050]

    gross, captured, net_list, ppm_series = [], [], [], []                # containers to fill
    g = scn.gross_start_mt                                                # current gross. Whatever I input to the function will be checked to see if it is a Scenario Object and take the current gross contained in that Scenario
    c = scn.ccus_start_mt                                                 # current captured
    decline = 1.0 - (scn.gross_decline_rate_pct / 100.0)                  # gross multiplier per year
    growth  = 1.0 + (scn.ccus_growth_rate_pct  / 100.0)                   # CCUS multiplier per year

    ppm_now = float(scn.starting_ppm)                                     # note: current atmospheric CO2 (ppm)

    for i, _ in enumerate(years):
        if i > 0:                                                         # note: apply year-to-year updates after first year
            g *= decline
            c *= growth

        gross.append(g)                                                   # note: record gross for this year
        captured.append(c)                                                # note: record captured for this year

        net_now = g - c                                                   # note: net emissions for this year (Mt)
        net_list.append(net_now)

        #UPDATE PPM
        add_ppm = (net_now / 1000.0) * scn.ppm_per_GtCO2                  # note: Mt -> Gt -> ppm
        excess  = ppm_now - scn.preindustrial_ppm                         # note: ppm above baseline
        sink    = scn.k_sink * excess                                     # note: removed by sinks this year
        ppm_now = ppm_now + add_ppm - sink                                # note: new atmospheric ppm
        ppm_series.append(ppm_now)                                        # note: store it

    # --- assemble results table ---
    df = pd.DataFrame({
        "Year": years,
        "Gross_Mt": np.array(gross, dtype=float),
        "Captured_Mt": np.array(captured, dtype=float),
        "Net_Mt": np.array(net_list, dtype=float),
        "Atmospheric_CO2_ppm": np.array(ppm_series, dtype=float),
    })
    return df

# RUN SCENARIO
default_scn = Scenario()                      # uses default scenario values
df_default = simulate_emissions(default_scn)  # Call function

# --- plot: emissions (left y) + ppm (right y) on the same figure ---
fig, ax1 = plt.subplots(figsize=(8,5))        # note: create figure + left axis
ax1.plot(df_default["Year"], df_default["Gross_Mt"], label="Gross (Mt)")
ax1.plot(df_default["Year"], df_default["Captured_Mt"], label="Captured (Mt)")
ax1.plot(df_default["Year"], df_default["Net_Mt"], label="Net (Mt)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Emissions (Mt CO₂/yr)")
ax1.grid(True)

ax2 = ax1.twinx()                             # note: right axis sharing the same x
ax2.plot(df_default["Year"], df_default["Atmospheric_CO2_ppm"], label="CO₂ (ppm)")
ax2.set_ylabel("CO₂ (ppm)")

# combine legends from both axes
l1, lab1 = ax1.get_legend_handles_labels()
l2, lab2 = ax2.get_legend_handles_labels()
ax1.legend(l1 + l2, lab1 + lab2, loc="upper right")

plt.title("Emissions & Atmospheric CO₂ (ppm) — Default Scenario")
plt.show()

# quick peek at the table
df_default.head()
```
